```python
# Once the loss function is defined, we calculate its derivative with respect to the model parameters.
# Using this derivative, we iteratively adjust the parameters so the loss function approaches its minimum value.
# This process is known as iterative optimization, where an optimizer is used to perform these parameter updates.

x = torch.autograd.Variable(torch.Tensor([2]), requires_grad=True)
y = torch.autograd.Variable(torch.Tensor([1]), requires_grad=True)
z = torch.autograd.Variable(torch.Tensor([5]), requires_grad=True)

a = x - y
f = z * a
print("Final value for Function f = ", f) # loss function
f.backward()
print("Gradient value for x:", x.grad)  # tensor([5.])
print("Gradient value for y:", y.grad)  # tensor([-5.])
print("Gradient value for z:", z.grad)  # tensor([1.])


```

## Model development and training

### Import libraries

```python
# Import libraries and dataset
from torchvision import models
from torch import optim
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

## Define neural network

```python
# Define neural network, __init__ and forward functions
class Net(nn.Module):                          # Define a custom neural network by extending nn.Module
    def __init__(self):                        # Constructor: define all layers here
        super(Net, self).__init__()            # Call the parent constructor (nn.Module)

        # First convolution layer: input has 3 channels (RGB), output has 6 feature maps, kernel size = 5x5
        self.conv1 = nn.Conv2d(3, 6, 5)

        # Max pooling layer: 2x2 filter, stride=2 (reduces width and height by half each time)
        self.pool = nn.MaxPool2d(2, 2)

        # Second convolution layer: input has 6 channels, output has 16 feature maps, kernel size = 5x5
        self.conv2 = nn.Conv2d(6, 16, 5)

        # First fully connected layer (dense): input features = 16*5*5, output = 120 neurons
        #nn.Linear(in_features, out_features) - in_features is number of inputs each neuron in this layer receives
        #out_features = the number of neurons in that layer (the outputs)
        self.fc1 = nn.Linear(16*5*5, 120)

        # Second fully connected layer: input = 120 neurons, output = 84 neurons
        self.fc2 = nn.Linear(120, 84)

        # Final fully connected layer: input = 84 neurons, output = 10 (for CIFAR-10 classes)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):                      # Forward pass: defines how input flows through the network
        # Apply first convolution -> ReLU activation -> Max pooling
        x = self.pool(F.relu(self.conv1(x)))

        # Apply second convolution -> ReLU activation -> Max pooling
        x = self.pool(F.relu(self.conv2(x)))

        # Flatten the 3D feature maps into a 1D vector for fully connected layers
        x = x.view(-1, 16 * 5 * 5)

        # Pass through first fully connected layer with ReLU
        x = F.relu(self.fc1(x)) # with relu all neagtive value set to zero and postive value left unchanged

        # Pass through second fully connected layer with ReLU
        x = F.relu(self.fc2(x))

        # Pass through final fully connected layer (no activation, raw class scores / logits)
        x = self.fc3(x)

        return x                               # Output: logits for 10 classes
```

## Instantiate the Model

```python
# Instantiate the model
net = Net()

# Define the loss function (CrossEntropyLoss is common for classification)
criterion = nn.CrossEntropyLoss()

# Define the optimizer (Stochastic Gradient Descent with momentum)
# lr=0.001 → how big each step is
# momentum=0.9 → how much of the previous velocity is kept
# 0.9 means 90% of the previous update is carried forward.
# This is a typical default in many models.
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9) # With momentum, the optimizer builds up speed in consistent directions
```

## Load and transform data

```python
import torchvision
# Define transformations: convert to tensor + normalize RGB channels
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(
        mean=(0.5, 0.5, 0.5),   # normalize each channel (R,G,B)
        std=(0.5, 0.5, 0.5)
    )
])

# Training dataset + dataloader
trainset = torchvision.datasets.CIFAR10(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

trainloader = torch.utils.data.DataLoader(
    trainset,
    batch_size=4,
    shuffle=True,
    num_workers=2
)

# Test dataset + dataloader
testset = torchvision.datasets.CIFAR10(
    root='./data',
    train=False,
    download=True,
    transform=transform
)

testloader = torch.utils.data.DataLoader(
    testset,
    batch_size=4,
    shuffle=False,
    num_workers=2
)
```

## Train the network

```python
# Train the Network

num_epochs = 10  # number of times to loop through the dataset

for epoch in range(num_epochs):
    running_loss = 0.0 #keeps track of the cumulative loss during the epoch

    for i, data in enumerate(trainloader, 0): #batch_idx -label for 1st batch is 0 #inputs.shape = [4, 3, 32, 32]  -4 images in the batch and labels of those images
        inputs, labels = data
        optimizer.zero_grad() #Clears old gradients from the previous step

        # Forward pass
        outputs = net(inputs) #Passes the batch of images through your neural network. #outputs.shape = [4, 10]
        loss = criterion(outputs, labels)
        #criterion is your loss function (nn.CrossEntropyLoss).# It compares:# outputs (predicted class scores)# labels (true class indices)

        # Backward pass + optimize
        loss.backward() #computes the gradients of the loss w.r.t. every trainable parameter (weights & biases) using backpropagation
        optimizer.step() #Updates the network parameters using the computed gradients, optimizer takes each parameter (weight, bias) and adjusts it slightly in the direction that reduces the loss
        # Accumulate loss
        running_loss += loss.item() # training loss since we are iterating training loader

        # Print statistics every 2000 mini-batches
        if i % 2000 == 1999:  # print every 2000 mini-batches
            print(f"[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 2000:.3f}")
            running_loss = 0.0

print("Finished Training ")
```
